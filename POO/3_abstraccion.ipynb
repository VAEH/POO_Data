```python
class Lavadora:

    def __init__(self):
        pass

    def lavar(self, temperatura='caliente'):
        self._llenar_tanque(temperatura)
        self._lavar()
        self._enjuagar()
        self._centrifugar()
    
    def _llenar_tanque(self, temperatura):
        print(f'Llenando tanque con agua {temperatura} :3')

    def _lavar(self):
        print('Empezando la lavación :3')

    def _enjuagar(self):
        print('Haciendo la enjuagación')
    
    def _centrifugar(self):
        print('Haciendo la centrifugación :3')


lavadora = Lavadora()

lavadora.lavar('Fria')
```
